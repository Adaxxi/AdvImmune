```python
import numpy as np
import torch
import seaborn as sns
from matplotlib import pyplot as plt
import pandas as pd

from utils import *
from attack import *
from adv_immunity import *
torch.cuda.empty_cache()

np_load_old = np.load
np.aload = lambda *a,**k: np_load_old(*a, allow_pickle=True, **k)
%load_ext memory_profiler
seed = 15
alpha = 0.85
%memit
```

```python
# 载入数据
file_name = 'data/karate.npz'
graph = load_dataset(file_name=file_name)
ori_adj, ori_attr, labels = standardize(adj_matrix=graph['adj_matrix'],
                             attr_matrix=graph['attr_matrix'],
                             labels=graph['labels'])
adj = torch.from_numpy(ori_adj.todense()).float()

labels = labels.astype('int32')
# loaded = np.load('data/cora_pretrained_logits.npz')
# logits = loaded['logits']
logits = np.load('data/karate_logits.npy')
%memit
```

```python
xi = 0
x = [0, 2, 4, 6]
# strength = [11, 9 ,7, 5]  local_budget=deg-11+strength
deg = ori_adj.sum(1).A1.astype(np.int32)
local_budget = np.maximum(deg-x[xi] , 0)
fragile = get_fragile(adj=ori_adj, threat_model='rem')
n, nc = logits.shape
non_control =  torch.ones(n,n)

ori_ppr_changing = pagerank_adj_changing(
    ori_adj, alpha, fragile, non_control, local_budget, logits)
np.save('karate_deg-%d_ori_ppr_changing.npy'%x[xi], ori_ppr_changing)

ori_worst_margins = worst_margins_given_k_squared(ori_ppr_changing, labels, logits)
print('Original Ratio of certified all nodes: {:.4f}'.format((ori_worst_margins>0).mean()))
print('%.6f'%ori_worst_margins.mean())
```

```python
# meta immunity
%memit
con_budget = 3
# ori_ppr_changing = np.aload('my_cora/local_deg/ori_ppr_changing.npy').item()
# all_controlled = np.aload('my_citeseer/local_deg/adj_controlled_%d.npy'%con_budget).item()
con_budget_local = local_budget
n, nc = logits.shape
con_budget_local = np.maximum(deg-x[xi] , 0)
exceed_local = []
for exceed_i in range(n):
    if con_budget_local[exceed_i] == 0:
        exceed_local.append(exceed_i)
# print('Equals local budget:', exceed_local)
adj_controlled = grad_adv_immune(adj, ori_ppr_changing, exceed_local, con_budget, logits, labels, alpha, con_budget_local)
for idx in range(n):
    tmp = np.where(adj_controlled[idx] == 0)[0].shape[0]
    if tmp > con_budget_local[idx]:
        print(idx)
```

```python
con_src = np.where(adj_controlled==0)[0]
con_tgt = np.where(adj_controlled==0)[1]

for i in range(con_budget):
    adj_controlled[con_tgt[i],con_src[i]] = 0
    
```

```python

# np.save('my_reddit/one_local_%d%d/adj_controlled_%d.npy'%(x[i],x[i],con_budget), adj_controlled)
print('local_budget = deg -', x[xi])
ori_worst_margins = worst_margins_given_k_squared(ori_ppr_changing, labels, logits)
print('Original Ratio of certified all nodes: {:.4f}'.format((ori_worst_margins>0).mean()))
print('Original num of robust node:', (ori_worst_margins>0).sum())

# all_controlled = np.aload('output_deg-3/adj_controlled_%d.npy'%con_budget).item()
con_ppr_changing = pagerank_adj_changing(ori_adj, alpha, fragile, adj_controlled, local_budget, logits)
con_worst_margins = worst_margins_given_k_squared(con_ppr_changing, labels, logits)
print('After control {:d}: Ratio of certified all nodes: {:.4f}'.format(con_budget, (con_worst_margins>0).mean()))
print('num of robust node:', (con_worst_margins>0).sum())
print(con_worst_margins.mean())
```

```python
n = labels.shape[0]
ori = np.where(ori_worst_margins>0)[0]
con = np.where(con_worst_margins>0)[0]
ori_rob = np.zeros(n)
con_rob = np.zeros(n)
ori_rob[ori] = 1
con_rob[con] = 1
idx = np.arange(n)
source = ori_adj.nonzero()[0]
target = ori_adj.nonzero()[1]
change = np.zeros(n)
change[con] = 1
change[ori] = 0
```

```python
pos = newpos

pos[33] = np.array([-0.51,0.14])
pos[32] = np.array([-0.39,0.07])
pos[31] = np.array([-0.34, -0.23312907])
pos[27] = np.array([-0.16270702, -0.23210303])
pos[13][1] = 0.07
pos[0][1] = 0

def min_max(arr, ymin, ymax):

    y = [((ymax-ymin)*(a-min(arr))/(max(arr)-min(arr)) + ymin) for a in arr]
    
    return (np.array(y))

n = len(labels)
xy = np.asarray([pos[v] for v in range(n)])

ori_min = ori_worst_margins.min()
ori_max = ori_worst_margins.max()
con_min = con_worst_margins.min()
con_max = con_worst_margins.max()
```

```python
fig = plt.figure(figsize=(15,10))
new_posx = min_max(xy[:,0], 0.09, 0.98)
new_posy = min_max(xy[:,1], 0.06, 0.93)

for i in range(n):
    if ori_worst_margins[i] > 0:
        smin = 0
        smax = con_worst_margins[i]
        c1, c2 = 'crimson', 'crimson'
    elif con_worst_margins[i] < 0:
        smin = ori_worst_margins[i]
        smax = 0
        c1, c2 = 'pink', 'pink'
    else:
        if i in [21]:
            smin = ori_worst_margins[i]
            smax = 0.1
        elif i in [0,27,33]:
            smin = -0.1
            smax = con_worst_margins[i]
#         elif i == 32:
#             smin = -0.2
#             smax = con_worst_margins[i]
        elif i in [24,25,32]:
            smin = -0.2
            smax = 0.2
        c1, c2 = 'pink', 'crimson'
    left, bottom, width, height = new_posx[i]-0.04,new_posy[i]-0.01,0.017,0.03
    if i == 0:
        bottom = bottom - 0.05
        left = left + 0.025
    elif i == 1:
        bottom = bottom +0.01
    ax2 = fig.add_axes([left,bottom,width,height],ylim=(smin,smax))
    bars = ax2.bar(range(1), ori_worst_margins[i], color=c1) + \
            ax2.bar(range(1, 2), con_worst_margins[i],color=c2)
    patterns = ('', '//////')

    for bar, pattern in zip(bars, patterns):
        bar.set_hatch(pattern)
    
    ax2.set_xticks([])
    ax2.spines['top'].set_visible(False)
    ax2.spines['right'].set_visible(False)
    ax2.spines['bottom'].set_visible(False)
    ax2.spines['left'].set_visible(False)

ax = fig.add_axes([0.033,0.0,1.0,1.0])

nx.draw_networkx(G, pos, node_color=label_color, font_size=18,
                 edge_color=con_color, width=wid, node_size=1900, with_labels=True)
ax.axis(False)
plt.savefig(fname="figures/karate_starnum.eps")
plt.show()
```
