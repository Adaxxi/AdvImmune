```python
import numpy as np
import torch
import seaborn as sns
from matplotlib import pyplot as plt

from utils import *
from attack import *
from adv_immunity import *
from baseline import *
import csv
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
torch.cuda.empty_cache()

np_load_old = np.load
np.aload = lambda *a,**k: np_load_old(*a, allow_pickle=True, **k)
%load_ext memory_profiler
seed = 15
alpha = 0.85
```

```python
# 载入数据
file_name = 'data/cora.npz'
graph = load_dataset(file_name=file_name)
ori_adj, ori_attr, ori_labels = standardize(adj_matrix=graph['adj_matrix'],
                             attr_matrix=graph['attr_matrix'],
                             labels=graph['labels'])
adj = torch.from_numpy(ori_adj.todense()).float()
edge_num = adj.sum().item()
# loaded = np.load('data/cora_pretrained_logits.npz')
# logits = loaded['logits']
logits = np.load('data/my_cora_logits.npy')
labels = np.load('output_csv/cora_predict.npy')
%memit
```

```python
i = 0
x = [0, 2, 4, 6]
# strength = [11, 9 ,7, 5]  local_budget=deg-11+strength
deg = ori_adj.sum(1).A1.astype(np.int32)
local_budget = np.maximum(deg-0 , 0)
# 生成可扰动边
mode = 'rem'
fragile = get_fragile(adj=ori_adj, threat_model=mode)
n, nc = logits.shape
con_budget_local = np.maximum(deg-0 , 0)
```

```python
ori_ppr_changing = np.aload('my_cora/one_attacklocal_%d%d/ori_ppr_changing.npy'%(0,0)).item()
file = 'my_cora/baseline.csv'
```

## 边的重要性

### local：bridgeness

```python
# 从大到小
budget_arr = np.arange(0,0.055,0.005)
i = 0
robust_arr = []
robnum_arr = []
con_worst_margins_arr = []

print('local_budget = deg -', x[i])
ori_worst_margins = worst_margins_given_k_squared(ori_ppr_changing, labels, logits)
print('Original Ratio of certified all nodes: {:.6f}'.format((ori_worst_margins>0).mean()))
print('Original num of robust node:', (ori_worst_margins>0).sum())
print(ori_worst_margins.mean())
robust_arr.append((ori_worst_margins>0).mean())
con_worst_margins_arr.append(ori_worst_margins.mean())

big_edge_neighb = edge_neighb_bridgeness(ori_adj, True)

# 各类别控制相同数量的边
cur_i = 0
cur_controlled = torch.ones(n,n)
for con_ratio in budget_arr[:-1]+0.005:
    con_budget = int(round(edge_num * con_ratio))
    diff = con_budget - int(round(edge_num * (con_ratio-0.005)))
    adj_controlled, cur_i = immune_edge_control_baseline(cur_controlled, big_edge_neighb, con_budget_local, cur_i, con_budget)
    con_ppr_changing = pagerank_adj_changing(ori_adj, alpha, fragile, adj_controlled, local_budget, logits)
    for ni in range(n):
        if np.where(adj_controlled[ni]!=1)[0].shape[0] > con_budget_local[ni]:
            print(ni)
    print('actual: ', np.where(adj_controlled!=1)[0].shape[0])
    con_worst_margins = worst_margins_given_k_squared(con_ppr_changing, labels, logits)
    print('After control {:d}: Ratio of certified all nodes: {:.6f}'.format(con_budget, (con_worst_margins>0).mean()))
    print('%.6f' % con_worst_margins.mean())
    cur_controlled = adj_controlled
    robust_arr.append((con_worst_margins>0).mean())
    con_worst_margins_arr.append(con_worst_margins.mean())

robust_arr = np.array(robust_arr)
con_worst_margins_arr = np.array(con_worst_margins_arr)

with open(file, 'w') as f:
    writer = csv.writer(f)
    writer.writerow(['bridgeness'])
dataframe = pd.DataFrame({u'控制边数':budget_arr,u'鲁棒节点比例':robust_arr,u'边距':con_worst_margins_arr})

dataframe.to_csv(file, mode='a')
    
```

```python
# 从大到小
budget_arr = np.arange(0,0.055,0.005)
i = 0
robust_arr = []
robnum_arr = []
con_worst_margins_arr = []
print('local_budget = deg -', x[i])
ori_worst_margins = worst_margins_given_k_squared(ori_ppr_changing, labels, logits)
print('Original Ratio of certified all nodes: {:.6f}'.format((ori_worst_margins>0).mean()))
print('Original num of robust node:', (ori_worst_margins>0).sum())
print(ori_worst_margins.mean())
robust_arr.append((ori_worst_margins>0).mean())
con_worst_margins_arr.append(ori_worst_margins.mean())
con_budget_local = np.maximum(deg-0 , 0)

similarity = 'Jaccard'
simil_matrix, _ = similarity_matrix(ori_adj, ori_attr, labels, similarity,mode)
cur_controlled = torch.ones((n,n))
for con_ratio in budget_arr[:-1]+0.005:
    con_budget = int(round(edge_num * con_ratio))
    diff = con_budget - int(round(edge_num * con_ratio-0.005))
    print(con_budget)
    adj_controlled, simil_matrix =  ICEU(ori_adj, cur_controlled, simil_matrix, _, con_budget, con_budget_local,mode)
    con_ppr_changing = pagerank_adj_changing(ori_adj, alpha, fragile, adj_controlled, local_budget, logits)
    for ni in range(n):
        if np.where(adj_controlled[ni]!=1)[0].shape[0] > con_budget_local[ni]:
            print(ni)
    print('actual: ', np.where(adj_controlled!=1)[0].shape[0])
    con_worst_margins = worst_margins_given_k_squared(con_ppr_changing, labels, logits)
    print('After control {:d}: Ratio of certified all nodes: {:.6f}'.format(con_budget, (con_worst_margins>0).mean()))
    print(con_worst_margins.mean())
    cur_controlled = adj_controlled
    robust_arr.append((con_worst_margins>0).mean())
    con_worst_margins_arr.append(con_worst_margins.mean())

robust_arr = np.array(robust_arr)
con_worst_margins_arr = np.array(con_worst_margins_arr)

with open(file, 'a') as f:
    writer = csv.writer(f)
    writer.writerow(['Jaccard'])
dataframe = pd.DataFrame({u'控制边数':budget_arr,u'鲁棒节点比例':robust_arr,
                          u'边距':con_worst_margins_arr})

dataframe.to_csv(file, mode='a')
    
```

## random

```python
# 基于普通的随机控制
budget_arr = np.arange(0,0.055,0.005)
robust_arr = []
con_worst_margins_arr = []
print('local_budget = deg -', x[i])
ori_worst_margins = worst_margins_given_k_squared(ori_ppr_changing, labels, logits)
print('Original Ratio of certified all nodes: {:.6f}'.format((ori_worst_margins>0).mean()))
print('Original num of robust node:', (ori_worst_margins>0).sum())
print(ori_worst_margins.mean())
robust_arr.append((ori_worst_margins>0).mean())
con_worst_margins_arr.append(ori_worst_margins.mean())
normal_fragile = np.column_stack(ori_adj.nonzero())
for con_ratio in budget_arr[:-1]+0.005:
    con_budget = int(round(edge_num * con_ratio))
    print(con_budget)
    rand_num = 10
    rob_ratio_all = []
    con_norm_all = []
    for i in range(rand_num):
        adj_controlled = random_control(normal_fragile, con_budget, con_budget_local,n,mode)
        for ni in range(n):
            if np.where(adj_controlled[ni]!=1)[0].shape[0] > con_budget_local[ni]:
                print('%d node is more than control local budget' %ni)
                print(ni)
        con_ppr_changing = pagerank_adj_changing(ori_adj, alpha, fragile, adj_controlled, local_budget, logits)
        con_worst_margins = worst_margins_given_k_squared(con_ppr_changing, labels, logits)
        rob_ratio_all.append((con_worst_margins>0).mean())
        con_norm_all.append(con_worst_margins.mean())
        del adj_controlled

    con_norm_all = np.array(con_norm_all)
    rob_ratio_all = np.array(rob_ratio_all)
    print('Ratio of robust node: %.6f' % rob_ratio_all.mean())
    print('Variance:%.6f' % rob_ratio_all.var())
    print('Worst margin: %.6f' % con_norm_all.mean())
    print('Variance:%.6f' % con_norm_all.var())
    robust_arr.append(con_norm_all.mean())
    con_worst_margins_arr.append(rob_ratio_all.mean())

robust_arr = np.array(robust_arr)
con_worst_margins_arr = np.array(con_worst_margins_arr)
with open(file, 'a') as f:
    writer = csv.writer(f)
    writer.writerow(['Random'])
dataframe = pd.DataFrame({u'控制边数':budget_arr,u'鲁棒节点比例':robust_arr,
                          u'边距':con_worst_margins_arr})
dataframe.to_csv(file, mode='a')
```
