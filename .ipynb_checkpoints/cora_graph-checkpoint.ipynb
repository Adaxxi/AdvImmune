```python
def similarity_matrix_(ori_adj, attr, labels, similarity='cosine', mode='add_rem'):
    if similarity == 'cosine':
        # 相似度矩阵越大，说明属性越相似
        simil_matrix = metrics.pairwise.cosine_similarity(attr, attr)
    elif similarity == 'Jaccard':
        # 距离矩阵越小，说明属性越相似
        distance_matrix = metrics.pairwise_distances(attr.todense(), metric='jaccard', n_jobs=10)
        simil_matrix = 1 - distance_matrix
    n = attr.shape[0]
    nc = labels.max()+1
    same_label = np.zeros((n,n))
    for c in range(nc):
        tmp = np.where(labels==c)[0]
        for t1 in tmp:
            same_label[t1,tmp] = 1
    connect_simil_matrix = np.multiply(simil_matrix,  ori_adj.todense())
    return connect_simil_matrix
```
